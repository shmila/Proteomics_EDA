```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
def create_protein_analysis_dataframe(csv_file, norm_type):
    df = pd.read_csv(csv_file)
    results = []
    
    for _, row in tqdm(df.iterrows(), total=len(df), desc='Analyzing Proteins'):
        identifiers = row.iloc[:4]
        
        # Find measurement columns
        if norm_type == 'LFQ':
            lysis_columns = [col for col in df.columns if col.startswith('LFQ intensity') and col.endswith('L')]
        else:
            lysis_columns = [col for col in df.columns if col.startswith(norm_type) and col.endswith('L')]
        
        patient_measurements = {}
        patient_cvs = {}
        
        for col in lysis_columns:
            measurement = row[col]
            
            # Extract patient number based on normalization type
            if norm_type == 'LFQ':
                patient = col.split(' ')[-1].split('_')[0]
            else:
                patient = col.split(' ')[1].split('_')[0]
            
            if patient not in patient_measurements:
                patient_measurements[patient] = []
            
            if measurement > 0:  # Only include non-zero measurements
                patient_measurements[patient].append(measurement)
        
        # Calculate CVs for patients with multiple measurements
        for patient, measurements in patient_measurements.items():
            if len(measurements) > 2:  # Require at least 3 measurements
                cv = np.std(measurements) / np.mean(measurements)
                patient_cvs[patient] = cv
        
        avg_cv = np.mean(list(patient_cvs.values())) if patient_cvs else np.nan
        
        results.append({
            **dict(zip(df.columns[:4], identifiers)),
            'Patient Measurements': patient_measurements,
            'Patient CVs': patient_cvs,
            'Average CV': avg_cv
        })
    
    return pd.DataFrame(results)
```

```python
for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    csv_file = f'relevant_patients_proteomics_table_{norm_type}.csv'
    result_df = create_protein_analysis_dataframe(csv_file, norm_type)
    result_df.to_csv(f'{norm_type}_protein_analysis.csv', index=False)
    print(f"{norm_type} analysis complete. Shape: {result_df.shape}")
```

```python
def inspect_data(csv_file):
    df = pd.read_csv(csv_file)
    df['Patient CVs'] = df['Patient CVs'].apply(eval)
    df['Patient Count'] = df['Patient CVs'].apply(len)
    print(f"\nFile: {csv_file}")
    print(f"Total rows: {len(df)}")
    print(f"Rows with Patient Count >= 1: {len(df[df['Patient Count'] >= 1])}")
    print(f"Average CV null values: {df['Average CV'].isna().sum()}")
    print("Sample of first row Patient CVs:", dict(df.iloc[0]['Patient CVs']))

for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    inspect_data(f'{norm_type}_protein_analysis.csv')
```

```python
def debug_csv_content(csv_file):
    df = pd.read_csv(csv_file)
    print(f"\nFile: {csv_file}")
    print("Columns:", df.columns.tolist())
    print("\nSample row:")
    print(df.iloc[0])

for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    debug_csv_content(f'relevant_patients_proteomics_table_{norm_type}.csv')
```

```python
def select_top_20_proteins(csv_file, min_patients=15):  
    # Load the protein analysis CSV
    df = pd.read_csv(csv_file)
    
    # Convert string representations of dicts to actual dicts
    df['Patient CVs'] = df['Patient CVs'].apply(eval)
    
    # Add patient count column
    df['Patient Count'] = df['Patient CVs'].apply(len)
    
    # Filter proteins with minimum number of patients
    filtered_df = df[df['Patient Count'] >= min_patients]
    
    # Sort by Average CV and select top 20
    top_20_proteins = filtered_df.sort_values('Average CV').head(20)
    
    return top_20_proteins
```

```python
for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    csv_file = f'{norm_type}_protein_analysis.csv'
    top_20_df = select_top_20_proteins(csv_file)
    top_20_df.to_csv(f'top_20_proteins_{norm_type}.csv', index=False)
    print(f"Top 20 proteins for {norm_type}: {top_20_df.shape[0]}")
```

```python
def convert_cv_values(input_file, output_file):
    df = pd.read_csv(input_file)
    df['Patient CVs'] = df['Patient CVs'].apply(eval)
    df['Patient CVs'] = df['Patient CVs'].apply(lambda d: {k: v/100 for k, v in d.items()})
    df.to_csv(output_file, index=False)

for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    input_file = f'top_20_proteins_{norm_type}.csv'
    output_file = f'top_20_proteins_{norm_type}_raw_cv.csv'
    convert_cv_values(input_file, output_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
intensity_df = pd.read_csv('top_20_proteins_Intensity_raw_cv.csv')
ibaq_df = pd.read_csv('top_20_proteins_iBAQ_raw_cv.csv')
lfq_df = pd.read_csv('top_20_proteins_LFQ_raw_cv.csv')

# Create figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('CV Distribution Across Normalization Methods')

# Plot for each method
for ax, (df, title) in zip(axes, [(intensity_df, 'Intensity'), 
                                 (ibaq_df, 'iBAQ'),
                                 (lfq_df, 'LFQ')]):
    sns.boxplot(y=df['Average CV'], ax=ax)
    sns.swarmplot(y=df['Average CV'], color='red', size=4, ax=ax)
    ax.set_title(f'{title} Method')
    ax.set_ylabel('CV')

plt.tight_layout()
plt.savefig('cv_distributions.png')
plt.close()

# Create violin plots for comparison
plt.figure(figsize=(10, 6))
data = []
labels = []

for df, method in [(intensity_df, 'Intensity'), 
                   (ibaq_df, 'iBAQ'),
                   (lfq_df, 'LFQ')]:
    data.extend(df['Average CV'].values)
    labels.extend([method] * len(df))

sns.violinplot(x=labels, y=data)
plt.title('CV Distribution Comparison')
plt.ylabel('Coefficient of Variation')
plt.savefig('cv_comparison.png')
plt.close()
```
