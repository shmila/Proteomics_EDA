```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read tumor data
intensity_df = pd.read_csv('tumor_top_20_proteins_Intensity.csv')
ibaq_df = pd.read_csv('tumor_top_20_proteins_iBAQ.csv')
lfq_df = pd.read_csv('tumor_top_20_proteins_LFQ.csv')
```

```python
# Distribution plots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('CV Distribution Across Normalization Methods (Tumor Samples)')

for ax, (df, title) in zip(axes, [(intensity_df, 'Intensity'), 
                                 (ibaq_df, 'iBAQ'),
                                 (lfq_df, 'LFQ')]):
    sns.boxplot(y=df['Average CV'], ax=ax)
    sns.swarmplot(y=df['Average CV'], color='red', size=4, ax=ax)
    ax.set_title(f'{title} Method')
    ax.set_ylabel('CV')

plt.tight_layout()
plt.savefig('tumor_cv_distributions.png')
plt.close()

# Violin plot
plt.figure(figsize=(10, 6))
data = []
labels = []
for df, method in [(intensity_df, 'Intensity'), 
                   (ibaq_df, 'iBAQ'),
                   (lfq_df, 'LFQ')]:
    data.extend(df['Average CV'].values)
    labels.extend([method] * len(df))

sns.violinplot(x=labels, y=data)
plt.title('CV Distribution Comparison (Tumor Samples)')
plt.ylabel('Coefficient of Variation')
plt.savefig('tumor_cv_comparison.png')
plt.close()
```
