```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_cv_distribution_comparison(norm_type):
    plt.figure(figsize=(12, 6))

    df = pd.read_csv(f'tumor_protein_analysis_{norm_type}.csv')
    top_20_df = pd.read_csv(f'tumor_top_20_proteins_{norm_type}.csv')
    df = df[df['Average CV'].notna()]

    non_top_20_cv = df[~df['Protein IDs'].isin(top_20_df['Protein IDs'])]['Average CV']

    plt.hist(non_top_20_cv, bins=5800, color='blue', alpha=0.7, 
            label=f'Other Proteins (n={len(non_top_20_cv)})')
    plt.hist(top_20_df['Average CV'], bins=20, color='red', alpha=0.7,
            label='Top 20 Most Stable Proteins')

    plt.xlabel('Average Coefficient of Variation (CV)')
    plt.ylabel('Number of Proteins')
    plt.title(f'Distribution of Protein Expression Variability\n{norm_type} Normalization')

    # Move legend outside plot
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    median_other = non_top_20_cv.median()
    median_top20 = top_20_df['Average CV'].median()
    stats_text = f'Median CV:\nTop 20: {median_top20:.3f}\nOthers: {median_other:.3f}'

    # Adjust text position
    plt.text(0.7, 0.95, stats_text,
            transform=plt.gca().transAxes,
            verticalalignment='top',
            horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.savefig(f'protein_cv_distribution_{norm_type}.png', dpi=300, bbox_inches='tight')
    plt.close()

for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    plot_cv_distribution_comparison(norm_type)
```

```python
def plot_cv_distribution_comparison(norm_type):
    plt.figure(figsize=(12, 6))

    # Read both files
    df = pd.read_csv(f'tumor_protein_analysis_{norm_type}.csv')
    top_20_df = pd.read_csv(f'tumor_top_20_proteins_{norm_type}.csv')

    # Filter for proteins with at least 2 patient measurements
    df['Patient CVs'] = df['Patient CVs'].apply(eval)
    df = df[df['Patient CVs'].apply(len) >= 2]
    df = df[df['Average CV'].notna()]

    # Get non-top-20 proteins
    non_top_20_cv = df[~df['Protein IDs'].isin(top_20_df['Protein IDs'])]['Average CV']

    # Plot histogram for other proteins
    plt.hist(non_top_20_cv, bins=5800, color='blue', alpha=0.7, 
            label=f'Other Proteins (n={len(non_top_20_cv)})')

    # Plot individual bars for top 20 proteins
    bar_height = max(plt.gca().get_ylim()) * 0.1  # Scale bar height to 10% of histogram height
    plt.bar(top_20_df['Average CV'], 
           height=bar_height,
           color='red', alpha=0.7,
           width=0.001,
           label='Top 20 Most Stable Proteins')

    plt.xlabel('Average Coefficient of Variation (CV)')
    plt.ylabel('Number of Proteins')
    plt.title(f'Distribution of Protein Expression Variability\n{norm_type} Normalization')

    # Move legend outside plot
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    # Calculate and display median values
    median_other = non_top_20_cv.median()
    median_top20 = top_20_df['Average CV'].median()
    stats_text = f'Median CV:\nTop 20: {median_top20:.3f}\nOthers: {median_other:.3f}'

    # Add statistics text
    plt.text(0.7, 0.95, stats_text,
            transform=plt.gca().transAxes,
            verticalalignment='top',
            horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.savefig(f'protein_cv_distribution_{norm_type}.png', dpi=300, bbox_inches='tight')
    plt.close()

# Generate plots for each normalization method
for norm_type in ['Intensity', 'iBAQ', 'LFQ']:
    plot_cv_distribution_comparison(norm_type)
```
